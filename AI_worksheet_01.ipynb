Problem - 3: Array Operations:


```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# Task 1: Add the two arrays
add_xy = x + y
print("Addition of x and y:\n", add_xy)

# Task 2: Subtract the two arrays
sub_xy = x - y
print("\nSubtraction of x and y:\n", sub_xy)

# Task 3: Multiply the array with an integer (example: 3)
mul_x = x * 3
print("\nMultiplication of x by 3:\n", mul_x)

# Task 4: Square of each element
square_x = x ** 2
print("\nSquare of elements in x:\n", square_x)

# Task 5: Dot product
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\nDot product of v and w:", dot_vw)
print("Dot product of x and v:\n", dot_xv)
print("Dot product of x and y:\n", dot_xy)

# Task 6: Concatenate x and y along rows (axis=0) and v and w along columns
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.column_stack((v, w))  # same as np.concatenate((v[:,None], w[:,None]), axis=1)
print("\nConcatenate x and y along rows:\n", concat_xy_row)
print("Concatenate v and w along columns:\n", concat_vw_col)

# Task 7: Concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    concat_xv_error = e
print("\nAttempt to concatenate x and v gives error:\n", concat_xv_error)
print("Explanation: x is 2x2 (2D) and v is 1D. Arrays must have the same number of dimensions along concatenation axis.")
```

Problem 4: Matrix Operations

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

# Task 1: Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("A * A^-1:\n", identity)
print("Is identity matrix? ", np.allclose(identity, np.eye(2)))  # np.allclose handles floating point precision

# Task 2: Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("BA:\n", BA)
print("AB equals BA? ", np.array_equal(AB, BA))

# Task 3: Prove (AB)^T = B^T * A^T
AB_T = (np.dot(A, B)).T
B_T_A_T = np.dot(B.T, A.T)
print("\n(AB)^T:\n", AB_T)
print("B^T * A^T:\n", B_T_A_T)
print("Check equality: ", np.array_equal(AB_T, B_T_A_T))

# Task 4: Solve the system of linear equations using inverse method
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Representing in matrix form: AX = B
coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
constant_matrix = np.array([-1, -3, 9])

# Solve using np.linalg.inv
X = np.dot(np.linalg.inv(coeff_matrix), constant_matrix)
print("\nSolution using inverse method:\n", X)

# Optional: Solve using np.linalg.solve (recommended)
X_solve = np.linalg.solve(coeff_matrix, constant_matrix)
print("Solution using np.linalg.solve:\n", X_solve)
```

```python
import numpy as np
import time

# -------------------------------
# 3. Dot Product
# -------------------------------
N = 1_000_000

# Using Python Lists
list1 = list(range(N))
list2 = list(range(N))

start_time = time.time()
dot_product_list = sum([list1[i] * list2[i] for i in range(N)])
end_time = time.time()
print("Dot Product using Python lists:", dot_product_list)
print("Time taken using Python lists:", end_time - start_time, "seconds")

# Using NumPy Arrays
array1 = np.arange(N)
array2 = np.arange(N)

start_time = time.time()
dot_product_array = np.dot(array1, array2)
end_time = time.time()
print("Dot Product using NumPy arrays:", dot_product_array)
print("Time taken using NumPy arrays:", end_time - start_time, "seconds")


# -------------------------------
# 4. Matrix Multiplication
# -------------------------------
size = 1000

# Using Python Lists
# Create matrices
matrix1 = [[i + j for j in range(size)] for i in range(size)]
matrix2 = [[i + j for j in range(size)] for i in range(size)]

start_time = time.time()
# Naive matrix multiplication
result_matrix = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(size)) for j in range(size)] for i in range(size)]
end_time = time.time()
print("\nMatrix multiplication using Python lists completed.")
print("Time taken using Python lists:", end_time - start_time, "seconds")

# Using NumPy Arrays
np_matrix1 = np.arange(size*size).reshape(size, size)
np_matrix2 = np.arange(size*size).reshape(size, size)

start_time = time.time()
np_result = np.dot(np_matrix1, np_matrix2)
end_time = time.time()
print("Matrix multiplication using NumPy arrays completed.")
print("Time taken using NumPy arrays:", end_time - start_time, "seconds")
```
